```python
import tensorflow as tf 
import keras
from keras.models import Sequential, Model
from keras.layers import Dense, Input
import numpy as np
import pandas as pd
import keras.backend as K
from sklearn import model_selection, preprocessing, metrics
# fix random seed for reproducibility
np.random.seed(7)
```

```python
df_text = pd.read_csv("words.csv")
```

```python
df_text.drop(columns = 'Unnamed: 0', inplace = True)
```

```python
df_text
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(df_text.iloc[:,0:2000], df_text.iloc[:,2000:2004], test_size = 0.2, random_state = 123)

# label encode the target variable
# encoder = preprocessing.LabelEncoder()
# y_train = encoder.fit_transform(y_train)
# y_test = encoder.fit_transform(y_test)
```

```python
y_train1 = y_train["total_total"]
y_train2 = y_train["E_score"]
y_train3 = y_train["S_score"]
y_train4 = y_train["G_score"]
```

```python
y_test1 = y_test["total_total"]
y_test2 = y_test["E_score"]
y_test3 = y_test["S_score"]
y_test4 = y_test["G_score"]
```

```python
# # create model
# model = tf.keras.models.Sequential()
# model.add(tf.keras.layers.Dense(12, input_dim=2000, activation='relu')) # 2000 12 8 4 model
# model.add(tf.keras.layers.Dense(8, activation='relu'))
# model.add(tf.keras.layers.Dense(4, activation='sigmoid'))

# Build a model
inputs = Input(shape = (2000,))
layer1 = Dense(64, activation = 'relu')(inputs)
layer2 = Dense(64, activation = 'relu')(layer1)
out1 = Dense(3, activation='linear')(layer2)
out2 = Dense(3, activation='linear')(layer2)
out3 = Dense(3, activation='linear')(layer2)
out4 = Dense(3, activation='linear')(layer2)
model = Model(inputs = inputs, outputs = [out1, out2, out3, out4])

# model.compile(optimizer = "rmsprop", loss = 'mse')
# predictions = Dense(3, activation = 'softmax')(layer2)
```

```python
# Compile the model
model.compile(optimizer = 'rmsprop',
              loss = 'sparse_categorical_crossentropy', # Call the loss function with the selected layer
              metrics = ['accuracy'])
```

```python
model.fit(x_train, [y_train1, y_train2, y_train3, y_train4], epochs = 5, batch_size = 8)
```

```python
val_loss, val_acc = model.evaluate(x_test, [y_test1, y_test2, y_test3, y_test4])
print(val_loss, val_acc)
```
